```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Bidirectional, GRU, Dense, Input, Attention, Dropout, Lambda
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import backend as K

# Define the model architecture
def create_bgru_model(input_shape, num_classes):
    # Input layer
    inputs = Input(shape=input_shape, name='input')

    # Bidirectional GRU Layer-1
    gru_layer1 = Bidirectional(GRU(256, return_sequences=True, dropout=0.2))(inputs)

    # Attention Layer
    attention = Attention()([gru_layer1, gru_layer1])

    # Apply softmax to attention weights
    attention_weights = Lambda(lambda x: K.softmax(x, axis=1))(attention)

    # Compute context vector
    context_vector = Lambda(lambda x: tf.reduce_sum(x[0] * x[1], axis=1))([attention_weights, gru_layer1])

    # Add an extra dimension to match expected input shape for the next GRU layer
    context_vector = Lambda(lambda x: tf.expand_dims(x, axis=1))(context_vector)

    # Bidirectional GRU Layer-2
    gru_layer2 = Bidirectional(GRU(256, return_sequences=False, dropout=0.2))(context_vector)

    # Fully Connected Layer
    fc = Dense(128, activation='relu')(gru_layer2)
    dropout = Dropout(0.5)(fc)  # Dropout layer for regularization
    output = Dense(num_classes, activation='softmax')(dropout)

    # Create and compile model
    model = Model(inputs=inputs, outputs=output)
    model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model

sequence_length = 18  # Update this to match the actual sequence length
num_features = 150  # Update this to match the actual number of features


num_classes = 12

# Create the model
# Update input_shape based on the above
input_shape = (sequence_length, num_features)
model = create_bgru_model(input_shape, num_classes)


# Print model summary
model.summary()

# Train the model (ensure X_train and y_train are defined in your environment)
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=40, batch_size=32)

# Evaluate the model (ensure X_test and y_test are defined in your environment)
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")

# Save model weights
# model.save_weights('path_to_pretrained_weights.h5')

# Load model weights (for transfer learning or evaluation)
# model.load_weights('path_to_pretrained_weights.h5')
```
